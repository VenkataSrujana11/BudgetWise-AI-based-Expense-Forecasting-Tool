```python
#Joining Numpy array using Vertical stack(Both arrays will be placed one after the other)
n10=np.array([1,2,3])
n11=np.array([4,5,6])
np.vstack((n10,n11))
```

```python
#Joining Numpy array using Horizontal stack(Both arrays will be placed side by side)
n12=np.array([1,2,3])
n13=np.array([4,5,6])
np.hstack((n12,n13))
```

```python
#Joining Numpy array using Column stack(Here each row will be converted into a column)
n14=np.array([1,2,3])
n15=np.array([4,5,6])
np.column_stack((n14,n15))
```

```python
#Intersection and Differences of Numpy arrays
n16=np.array([1,2,3,4,5])
n17=np.array([4,5,6,7])
np.intersect1d(n16,n17)
```

```python
#Gives elements in array A that are not in array B
np.setdiff1d(n16,n17)
```

```python
np.setdiff1d(n17,n16)
```

```python
#Creating a pie chart
fruit=['Apple','Banana','Orange','Mango']
quantity=[67,34,100,29]
plt.pie(quantity,labels=fruit)
plt.show()
```

```python
#Line plot
x=np.arange(1,11)
y1=2*x
y2=3*x
```

```python
#Line Plot example(Adding two lines in the same plot)
plt.plot(x,y1,color='g',linestyle=':',linewidth='2')
plt.plot(x,y2,color='r',linestyle='-.',linewidth='3')
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
#Adding sub-plots
y1=2*x
y2-3*x

plt.subplot(1,2,1)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)

plt.subplot(1,2,2)
plt.plot(x,y2,color='r',linestyle='-.',linewidth=2)

plt.show()
```

Categorical Plot

```python
#Categorical scatterplots
#The default representation of the data in catplot() uses a scatterplot
#There are actually two different categorical scatter plots in seaborn
import seaborn as sns
```

```python
#Loading tips dataset from seaborn and printing the first 5 rows
tips=sns.load_dataset('tips')
tips.head()
```

```python
#Plotting a categorical plot
sns.catplot(data=tips,x='day',y='total_bill')
```

```python
#The jitter parameter controls the magnitude of jitter or disables it altogether
sns.catplot(data=tips,x='day',y='total_bill',jitter=False)
```

```python
#Prevent from overlapping(Swarm plot)
sns.catplot(data=tips,x='day',y='total_bill',kind='swarm')
```

```python
#Add the hue semantic
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='swarm')
```

```python
#Order parameter: to display multiple categorical plot in the single figure
sns.catplot(data=tips,x='smoker',y='tip',order=['No','Yes'])
```

```python
#Categorical plot on vertical axis
sns.catplot(data=tips,x='total_bill',y='day',hue='time',kind='swarm')
```

```python
#Comparing Distributions
#Boxplots
sns.catplot(data=tips,x='day',y='total_bill',kind='box')
```

```python
#Adding hue semantic
sns.catplot(data=tips,x='day',y='total_bill',hue='smoker',kind='box')
```

```python
#VIOLIN plor
sns.catplot(data=tips,x='total_bill',y='day',hue='sex',kind='violin')
```

```python
#Split in the violin plot
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='violin',split=True)
```
